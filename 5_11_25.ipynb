```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# 1. URL de la API
url = "https://jsonplaceholder.typicode.com/users"

# 2. Solicitar datos a la API
response = requests.get(url)
data = response.json()

# 3. Convertir a DataFrame
df = pd.DataFrame(data)

# 4. Extraer ciudad (viene dentro del campo "address")
df['ciudad'] = df['address'].apply(lambda x: x['city'])

# 5. Contar usuarios por ciudad
conteo = df['ciudad'].value_counts()

# 6. Crear gráfica
plt.figure()
conteo.plot(kind='bar')   # No se especifican colores (regla para analítica)
plt.xlabel("Ciudad")
plt.ylabel("Número de Usuarios")
plt.title("Usuarios por Ciudad")
plt.show()
```
