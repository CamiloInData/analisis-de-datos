```python
import pandas as pd
import numpy as np

df = pd.read_csv("DatosVentas.csv")
print(df)
```

SE A;ADE LA COMUNA "UNIDADES DISPONIBLES" PARA PODER COMPLETAAR EL EJERCICIO , SE CAMBIA EL TIPO DE DATO DE PRECIO UNITARIO.

```python

df['Total Ventas'] = np.random.randint(2, 180, size=20)
df['PrecioUnitario'] = df['PrecioUnitario'].str.replace(',', '.').astype(float)

print(df)
```

Se muestran los productos que tienen un precio unitario superior al promedio y que además hayan vendido más de 40 unidades

```python

promedio_precio = df['PrecioUnitario'].mean()

resultado = df[(df['PrecioUnitario'] > promedio_precio) & (df['Total Ventas'] > 40)][['Producto', 'Categoría', 'PrecioUnitario', 'Total Ventas']]

print(f"Promedio de precio: {promedio_precio:.2f}")
print("\nProductos filtrados:")
print(resultado)
```
