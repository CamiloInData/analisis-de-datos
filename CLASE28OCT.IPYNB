EJERCICIO 1 :



```python
user = input("Ingresa números separados por espacios: ")
lista_texto = user.split()
numeros = []
for num_texto in lista_texto:
    numeros.append(int(num_texto))
```

EJERCICIO 2:



```python
suma = 0
for num in numeros:
    suma = suma + num
mayores = []
for num in numeros:
    if num > 10:
        mayores.append(num)
print("La suma de todos los números es:", suma)
print("Los números mayores a 10 son:", mayores)
```
