```python
# Preparaci√≥n
import pandas as pd
import numpy as np

print("Pandas:", pd.__version__, "| Numpy:", np.__version__)
datos = [
    ['Pan', 10, 2000],
    ['Leche', 20, 3500],
    ['Huevos', 30, 500],
    ['Arroz', 15, 2500],
    ['Caf√©', 8, 8000]
]

df = pd.DataFrame(datos, columns=['Producto', 'Cantidad', 'Precio'])
df
df['Total'] = df['Cantidad'] * df['Precio']
df['IVA'] = df['Total'] * 0.19
df['Total_con_IVA'] = df['Total'] + df['IVA']
df
alto_precio = df[df['Precio'] > 3000]
alto_total_iva = df[df['Total_con_IVA'] > 50000]

print("Precio > 3000:\n", alto_precio, "\n")
print("Total con IVA > 50_000:\n", alto_total_iva)

primera_fila = df.iloc[0]
filas_1_3 = df.iloc[1:4]
productos = df['Producto']

print("Primera fila:\n", primera_fila, "\n")
print("Filas 1 a 3:\n", filas_1_3, "\n")
print("Columna Producto:\n", productos)

fila_nueva = ['Az√∫car', 12, 1500]
df = pd.concat([df, pd.DataFrame([fila_nueva], columns=df.columns[:3])], ignore_index=True)

# volver a calcular Totales para la nueva fila
df['Total'] = df['Cantidad'] * df['Precio']
df['IVA'] = df['Total'] * 0.19
df['Total_con_IVA'] = df['Total'] + df['IVA']

df.reset_index(drop=True, inplace=True)
df

datos_con_nan = [
    ['Galleta', 5, 1200],
    ['Jugo', 7, np.nan],   # precio faltante
    ['Yogurt', 10, 2000]
]

df_nan = pd.DataFrame(datos_con_nan, columns=['Producto', 'Cantidad', 'Precio'])
df_nan['Precio'].isnull().sum(), df_nan
media_precio = df_nan['Precio'].mean(skipna=True)
df_nan['Precio'].fillna(media_precio, inplace=True)
df_nan['Total'] = df_nan['Cantidad'] * df_nan['Precio']
df_nan

df2 = pd.DataFrame([
    ['Pan', 'Abarrotes', 10, 2000],
    ['Arroz', 'Abarrotes', 15, 2500],
    ['Leche', 'L√°cteos', 20, 3500],
    ['Yogurt', 'L√°cteos', 10, 2000],
    ['Caf√©', 'Bebidas', 8, 8000]
], columns=['Producto', 'Categoria', 'Cantidad', 'Precio'])

df2['Total'] = df2['Cantidad'] * df2['Precio']

resumen = df2.groupby('Categoria')['Total'].agg(['sum','mean','count']).reset_index()
resumen
ordenado = df2.sort_values(by='Total', ascending=False)
ordenado.to_csv('productos_resumen.csv', index=False)
ordenado

```

```python
# Preparaci√≥n
import pandas as pd      # Librer√≠a principal para an√°lisis de datos
import numpy as np       # Librer√≠a para operaciones num√©ricas

# Muestra las versiones instaladas
print("Pandas:", pd.__version__, "| Numpy:", np.__version__)
```

Explicaci√≥n:
Estas librer√≠as permiten manipular datos en forma de tablas (DataFrames) y realizar c√°lculos matem√°ticos f√°cilmente.

2. Crear un DataFrame b√°sico

```python
# Lista de listas con los datos
datos = [
    ['Pan', 10, 2000],
    ['Leche', 20, 3500],
    ['Huevos', 30, 500],
    ['Arroz', 15, 2500],
    ['Caf√©', 8, 8000]
]

# Crear el DataFrame
df = pd.DataFrame(datos, columns=['Producto', 'Cantidad', 'Precio'])
df
```

Explicaci√≥n:
Cada fila representa un producto con su cantidad y precio.
El DataFrame se comporta como una ‚Äútabla de Excel‚Äù en Python.

3. Agregar nuevas columnas

```python
# Calcular el total por producto
df['Total'] = df['Cantidad'] * df['Precio']

# Calcular IVA (19%)
df['IVA'] = df['Total'] * 0.19

# Calcular total con IVA incluido
df['Total_con_IVA'] = df['Total'] + df['IVA']

df
```

Explicaci√≥n:
Creamos columnas calculadas. Pandas aplica las operaciones a toda la columna, sin necesidad de bucles.

4. Filtrar datos seg√∫n condiciones

```python
# Productos con precio mayor a 3000
alto_precio = df[df['Precio'] > 3000]

# Productos con Total + IVA mayor a 50,000
alto_total_iva = df[df['Total_con_IVA'] > 50000]

print("Precio > 3000:\n", alto_precio, "\n")
print("Total con IVA > 50_000:\n", alto_total_iva)
```

Explicaci√≥n:
Se filtran filas seg√∫n condiciones l√≥gicas, igual que ‚Äúfiltros‚Äù en Excel.

5. Acceso a filas y columnas

```python
# Primera fila del DataFrame
primera_fila = df.iloc[0]

# Filas desde la 1 hasta la 3
filas_1_3 = df.iloc[1:4]

# Acceder a una sola columna
productos = df['Producto']

print("Primera fila:\n", primera_fila, "\n")
print("Filas 1 a 3:\n", filas_1_3, "\n")
print("Columna Producto:\n", productos)
```

Explicaci√≥n:

iloc accede por posici√≥n num√©rica.

df['columna'] accede a una columna espec√≠fica.

6. Agregar una nueva fila

```python
# Nueva fila
fila_nueva = ['Az√∫car', 12, 1500]

# Agregarla al DataFrame original
df = pd.concat([df, pd.DataFrame([fila_nueva], columns=df.columns[:3])], ignore_index=True)

# Recalcular las columnas derivadas
df['Total'] = df['Cantidad'] * df['Precio']
df['IVA'] = df['Total'] * 0.19
df['Total_con_IVA'] = df['Total'] + df['IVA']

# Reiniciar el √≠ndice
df.reset_index(drop=True, inplace=True)
df
```

Explicaci√≥n:
Agregamos datos nuevos y actualizamos c√°lculos para mantener consistencia.

7. Manejo de valores faltantes (NaN)

```python
datos_con_nan = [
    ['Galleta', 5, 1200],
    ['Jugo', 7, np.nan],   # precio faltante
    ['Yogurt', 10, 2000]
]

df_nan = pd.DataFrame(datos_con_nan, columns=['Producto', 'Cantidad', 'Precio'])

# Contar valores faltantes
df_nan['Precio'].isnull().sum(), df_nan

# Calcular la media de los precios (sin contar NaN)
media_precio = df_nan['Precio'].mean(skipna=True)

# Rellenar el valor faltante con la media
df_nan['Precio'].fillna(media_precio, inplace=True)

# Crear columna Total
df_nan['Total'] = df_nan['Cantidad'] * df_nan['Precio']
df_nan
```

Explicaci√≥n:
Se identifican valores faltantes (NaN), se calcula la media y se rellenan autom√°ticamente.

8. Agrupar datos y obtener resumen estad√≠stico

```python
df2 = pd.DataFrame([
    ['Pan', 'Abarrotes', 10, 2000],
    ['Arroz', 'Abarrotes', 15, 2500],
    ['Leche', 'L√°cteos', 20, 3500],
    ['Yogurt', 'L√°cteos', 10, 2000],
    ['Caf√©', 'Bebidas', 8, 8000]
], columns=['Producto', 'Categoria', 'Cantidad', 'Precio'])

df2['Total'] = df2['Cantidad'] * df2['Precio']

# Agrupar por categor√≠a y obtener suma, promedio y cantidad
resumen = df2.groupby('Categoria')['Total'].agg(['sum', 'mean', 'count']).reset_index()
resumen
```

Explicaci√≥n:
groupby() permite analizar datos por grupos (por ejemplo, totales por categor√≠a).

9. Ordenar y exportar datos

```python
# Ordenar por el total de forma descendente
ordenado = df2.sort_values(by='Total', ascending=False)

# Exportar a archivo CSV
ordenado.to_csv('productos_resumen.csv', index=False)

ordenado
```

Explicaci√≥n:
Ordenamos la tabla y la guardamos como archivo .csv, √∫til para an√°lisis o compartir resultados.

¬øQu√© es None?

 None es una palabra clave de Python (no de Pandas).
Se usa para representar la ausencia de valor o un valor nulo.

 Es del tipo NoneType, y lo encuentras en Python puro, por ejemplo:

```python
x = None
print(x)         # None
print(type(x))   # <class 'NoneType'>
```

Ejemplo real:
Cuando una funci√≥n no devuelve nada, en realidad devuelve None.

¬øQu√© es NaN?

 NaN significa Not a Number (‚Äúno es un n√∫mero‚Äù).
Viene de NumPy (numpy.nan) y Pandas lo usa internamente para representar valores faltantes en datos num√©ricos.

```python
import numpy as np
y = np.nan
print(y)          # nan
print(type(y))    # <class 'float'>
```

NaN aparece mucho cuando trabajas con archivos CSV o c√°lculos que fallan (por ejemplo, dividir 0/0).

Diferencias
| Concepto                      | `None`                                 | `NaN`                              |
| ----------------------------- | -------------------------------------- | ---------------------------------- |
| **Origen**                    | Python                                 | NumPy / Pandas                     |
| **Significado**               | Ausencia total de valor                | Valor num√©rico faltante            |
| **Tipo de dato**              | `NoneType`                             | `float`                            |
| **Uso com√∫n**                 | En listas, funciones o texto           | En columnas num√©ricas o DataFrames |
| **Se detecta con `isnull()`** | ‚úÖ S√≠                                   | ‚úÖ S√≠                               |
| **Usado por Pandas**          | üî∏ Convertido autom√°ticamente en `NaN` | ‚úÖ Valor est√°ndar para ‚Äúfaltante‚Äù   |


¬øCu√°l se usa en Pandas?

 En Pandas, se usa NaN, no None.
Pandas convierte autom√°ticamente los None a NaN cuando creas un DataFrame.

```python
import pandas as pd
import numpy as np

datos = {
    'Producto': ['Pan', 'Leche', 'Jugo'],
    'Precio': [2000, None, np.nan]
}

df = pd.DataFrame(datos)
print(df)
```

Observa que el valor None fue convertido a NaN autom√°ticamente.
| Acci√≥n                    | Funci√≥n             | Ejemplo                                |
| ------------------------- | ------------------- | -------------------------------------- |

| Calcular promedio sin NaN | `mean(skipna=True)` | `df['Precio'].mean()`                  |
