```python
#Base de datos Titanic
import pandas as pd
import numpy as np
url="https://drive.google.com/uc?export=download&id=19yNKOtKt_bUZa8eNr5dKpoLE7lVVoIeO"
titanic=pd.read_csv(url)
titanic.head()
titanic.info()
titanic.shape
```

1.PRIMERA LIMPIEZA DE DATOS: [CABIN]

```python
titanic['TieneCabina'] = titanic['Cabin'].notna().astype(int)
titanic = titanic.drop('Cabin', axis=1)

titanic
```

Limpieza de dato age * nueva columna : se crea una nueva columna con los datos de el titulo incluido en el nombre para incluir valor , se recogen los titulos y se promedian las edades para rellenar valores nulos en la edad.

```python
# 1) Extraer el título usando expresión regular
titanic['Title'] = titanic['Name'].str.extract(r',\s*([^.]*)\.')

# 2) (Opcional) limpiar el punto final si deseas títulos sin punto
# En este método ya salen sin punto, pero si quieres asegurarlo:
titanic['Title'] = titanic['Title'].str.strip()

# 3) Ver las primeras filas para verificar
print(titanic[['Name', 'Title']].head())

# 4) Ver los títulos únicos que aparecen en el dataset
print(titanic['Title'].unique())

print (titanic.head)
```

```python
titanic
```

```python

titanic.loc[(titanic['Title'] == 'Master') & (titanic['Age'] > 18), 'Age'] = np.nan


promedioedad = titanic.groupby('Title')['Age'].mean()

# 3) Rellenar los valores faltantes usando ese promedio
titanic['Age'] = titanic['Age'].fillna(titanic['Title'].map(promedioedad))

# 4) Verifica los cambios
print(titanic[['Title', 'Age']].head())
print(titanic['Age'].isnull().sum()) 
```
