Crear una lista de listas con un valor faltante

```python
import pandas as pd 
import numpy as np
```

```python
datos_con_nan = [
    ['Galleta', 5, 1200],
    ['Jugo', 7, np.nan],   # precio faltante
    ['Yogurt', 10, 2000]
]
df_nan = pd.DataFrame(datos_con_nan, columns=['Producto', 'Cantidad', 'Precio'])
print(df_nan)
```

Deteccion de todos los valores nuloes :

```python
df_nan['Precio'].isnull().sum()
```

Calcular la media ignorando los NaN

```python
media_precio = df_nan['Precio'].mean(skipna=True)
print(media_precio)
```

Reemplaza valores de NaN con la media:

```python
df_nan['Precio'].fillna(media_precio, inplace=True)
print(df_nan)
```

Calcular la columna total :

```python
df_nan['Total'] = df_nan['Cantidad'] * df_nan['Precio']
print(df_nan)
```
