Las cadenas se utilizan para almacenar texto. Se pueden delimitar mediante comillas simples o dobles.

```python
#mostrar una cadena y su longitud
#ejercicio1
#ingresamos a una cadena
texto = input("Escribe una frase: ")
#mostramos el texto para saber cuantos caracteres tiene
print("Tu frase es:", texto)
print("Número de caracteres:", len(texto))
```

```python
#convertir de munusculas a mayusculas
#ejercicio 2
#ingresamos a una cadena
texto = input("Escribe una palabra: ")

print("En mayúsculas:", texto.upper())
print("En minúsculas:", texto.lower())
```

```python
#contar cuantas veces aparece una letra
frase = input("Escribe una frase: ")
letra = input("Que letra deseas contar: ")
print(f"La letra '{letra}' aparece", frase.count(letra), "veces.")
```

```python
#mostrar parte de la cadena

cadena = "Hola Mundo"
print("Los primeros 4 caracteres son:", cadena[0:4])
print("Los últimos 3 caracteres son:", cadena[-3:])
```

```python
#separar palabras y unirlas
frase = "Aprender Python es divertido"

palabras = frase.split()  # Separa por espacios
print("Lista de palabras:", palabras)

nueva_frase = "_".join(palabras)  # Une con guiones
print("Frase unida con guiones:", nueva_frase)
```

```python
#eliminar espacios
nombre = "   Laura  paola   . "
print("Antes:", nombre)
print("Después:", nombre.strip())
```

```python
# Verificar si una palabra está en la frase
frase = input("Escribe una frase: ")
palabra = input("Escribe una palabra: ")

if palabra in frase:
    print(f"La palabra '{palabra}' está en la frase.")
else:
    print(f"La palabra '{palabra}' NO está en la frase.")
```

## Condicionales

Los condicionales son una forma de ejecutar código dependiendo de si una condición es Verdadera o Falso. Al igual que en otros lenguajes, Python admite if y else, pero else if se contrae a elif, como lo demuestra el siguiente ejemplo.

```python
# Correr la siguiente linea 3 veces, con un número positivo, otro negativo y con 0
numero = int(input('Ingrese un número entre -10 y 10 '))
if numero < 0:
  print("negativo")
elif numero == 0:
  print("nulo")
else: # numero > 0
  print("positivo")
```

## Bucles

Los bucles son una forma de ejecutar un bloque de código varias veces. Hay dos tipos principales de bucles: bucles while y bucles for.

```python
lista = [1, 2, 3]
```

Usando while

```python
i = 0
while i < len(lista):
  print(lista[i])
  i += 1 # equivalente a i = i + 1
```

Usando for

```python
for i in range(len(lista)):
  print(lista[i])
```

De otra forma:

```python
for element in lista:
  print(element)
```
