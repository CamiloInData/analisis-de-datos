```python
# Escribir una función para calcular el número de letras minúsculas y mayúsculas tiene una frase
```

```python
def contar_minuscula_mayuscula(texto):
    contador_mayusculas = 0
    contador_minusculas = 0
    for letra in texto:  
        if letra.isupper():
            contador_mayusculas += 1
        elif letra.islower():
            contador_minusculas += 1
    return contador_mayusculas, contador_minusculas

texto = input("Escribe una frase: ")


mayor, menor = contar_minuscula_mayuscula(texto)


print(f"Mayúsculas: {mayor}")
print(f"Minúsculas: {menor}")
```
