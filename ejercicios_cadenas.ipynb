```python
#Ejercicio 1
# Solicita al usuario su nombre y apellido, y muestra un saludo completo

nombre = input("Escribe tu nombre: ")
apellido = input("Escribe tu apellido: ")

saludo = "Hola " + nombre + " " + apellido + ", ¡bienvenido!"
saludo = saludo.upper()
print(saludo)

# Desafío:
# Haz que todo el saludo aparezca en mayúsculas
#print(saludo.upper())
```

```python
# Muestra diferentes partes de una cadena

palabra = "Programación"
print("Primeras 4 letras:", palabra[0:4])
print("Últimas 3 letras:", palabra[-3:])
print("Desde la letra 3 hasta la 8:", palabra[2:8])

# Desafío:
# Mostrar la palabra al revés
print(palabra[::-1])
```

```python
# Aplicar métodos comunes a cadenas

frase = "   Python es divertido   "
print(frase.upper())   # Mayúsculas
print(frase.lower())   # Minúsculas
print(frase.title())   # Título
print(frase.strip())   # Quitar espacios
print(frase.replace("divertido", "increíble"))  # Reemplazar palabra

# Desafío:
# Cambiar dos palabras seguidas usando replace()
print(frase.replace("Python", "Java").replace("divertido", "potente"))
```

```python
# Verificar si una palabra está dentro de una frase

frase = input("Escribe una frase: ")
palabra = input("Escribe una palabra: ")

if palabra in frase:
    print(f"La palabra '{palabra}' está en la frase.")
else:
    print("La palabra NO está en la frase.")

# Desafío:
# Hacer que no distinga mayúsculas/minúsculas
# if palabra.lower() in frase.lower():
#     print("Está presente (sin importar mayúsculas/minúsculas)")
```

```python
# Contar cuántas veces aparece una palabra o letra en una frase

frase = "Me gusta programar en Python porque Python es genial"
print("La palabra 'Python' aparece", frase.count("Python"), "veces.")

# Desafío:
# Pide al usuario qué palabra quiere contar
palabra = input("¿Qué palabra quieres contar?: ")
print(f"La palabra '{palabra}' aparece {frase.count(palabra)} veces.")
```

```python
# Crear una frase personalizada con variables

nombre = "Ana"
edad = 20
print(f"{nombre} tiene {edad} años y estudia programación.")

# Desafío:
# Pide al usuario su nombre, edad y ciudad, y crea una frase
nombre = input("Nombre: ")
edad = input("Edad: ")
ciudad = input("Ciudad: ")
print(f"{nombre} tiene {edad} años y vive en {ciudad}.")
```

```python
# Comprobar si una frase comienza o termina con cierta palabra

frase = "Aprender Python es divertido"
print(frase.startswith("Aprender"))
print(frase.endswith("divertido"))

# Desafío:
# Pide al usuario la palabra de inicio y final
inicio = input("Palabra inicial: ")
final = input("Palabra final: ")
print(frase.startswith(inicio))
print(frase.endswith(final))
```
