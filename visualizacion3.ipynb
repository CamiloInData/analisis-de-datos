```python
# ========================================
# 1. Importaci칩n de librer칤as
# ========================================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo visual bonito
sns.set(style="whitegrid")

# ========================================
# 2. Cargar dataset "tips" (propinas)
# ========================================
df = sns.load_dataset("tips")
df.head()
```

```python
# ========================================
# 3. Exploraci칩n b치sica
# ========================================
print(df.info())
print("\nDescripci칩n estad칤stica:\n", df.describe())
print("\nColumnas del dataset:", df.columns.tolist())
```

```python
# ========================================
# 4. Gr치fico de barras
# Comparar propina promedio por d칤a
# ========================================
sns.barplot(x="day", y="tip", data=df, estimator="mean")
plt.title("Propina promedio por d칤a")
plt.show()
```

```python
# ========================================
# 5. Gr치fico de l칤neas
# Ver c칩mo cambia la cuenta total por tama침o del grupo
# ========================================
sns.lineplot(x="size", y="total_bill", data=df)
plt.title("Tendencia de la cuenta total seg칰n el tama침o del grupo")
plt.show()
```

```python
# ========================================
# 6. Gr치fico de dispersi칩n
# Relaci칩n entre cuenta total y propina
# ========================================
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)
plt.title("Relaci칩n entre cuenta total y propina")
plt.show()
```

```python
# ========================================
# 7. Boxplot (gr치fico de caja)
# Distribuci칩n de propinas por d칤a
# ========================================
sns.boxplot(x="day", y="tip", data=df)
plt.title("Distribuci칩n de propinas por d칤a")
plt.show()
```

```python
# ========================================
# 8. Histograma
# Distribuci칩n de la cuenta total
# ========================================
sns.histplot(df["total_bill"], bins=10, kde=True)
plt.title("Distribuci칩n de la cuenta total")
plt.show()
```

```python
# ========================================
# 9. Gr치fico circular (Pie Chart)
# Proporci칩n de clientes por sexo
# ========================================
df["sex"].value_counts().plot.pie(autopct='%1.1f%%', colors=["#66b3ff","#ff9999"])
plt.title("Proporci칩n de clientes por sexo")
plt.ylabel("")
plt.show()
```

```python
# ========================================
# 10. Mapa de calor (Heatmap)
# Correlaciones entre variables num칠ricas
# ========================================
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Mapa de calor de correlaciones num칠ricas")
plt.show()
```

```python
# ========================================
# 11. Gr치fico de viol칤n
# Distribuci칩n de propinas seg칰n fumadores
# ========================================
sns.violinplot(x="smoker", y="tip", data=df)
plt.title("Distribuci칩n de propinas seg칰n fumadores")
plt.show()
```
